We can now write this data to our data directory, and prepare to make an LCA calculation.

```python
biosphere = Database("simple-b")
if "simple-b" not in databases:
    biosphere.register(format="handmade", depends=[], num_processes=4)
biosphere.write(biosphere_data)
biosphere.process()
technosphere = Database("simple-t")
if "simple-t" not in databases:
    technosphere.register(format="handmade", depends=["simple-b"], num_processes=5)
technosphere.write(technosphere_data)
technosphere.process()
```

Of course, we also need an LCIA method, but the definition for this is much simpler:

```python
lcia_data = [
    [("simple-b", "I"), 0.1, "GLO"],
    [("simple-b", "II"), 0.2, "GLO"],
    [("simple-b", "III"), 0.3, "GLO"],
    [("simple-b", "IV"), 0.4, "GLO"]
    ]
lcia_name = ("simple", "method")
lcia_method = Method(lcia_name)
if lcia_name not in methods:
    lcia_method.register(unit="points")
lcia_method.write(lcia_data)
lcia_method.process()
```

Our "final product" activity is E, and for this example, let's take 2.72 units of E. We have to reference it by its full identifier:

```python
lca = LCA(demand={("simple-t", "E"): 2.72}, method=("simple", "method"))
lca.lci()  # Calculate inventory
lca.lcia()  # Calculate impact assessment
print lca.score
```

This isn't all that helpful, as this is just a made-up example. Let's make an online report so we can learn a little more:

```python
from bw2analyzer import SerializedLCAReport
report = SerializedLCAReport({("simple-t", "E"): 2.72}, method=("simple", "method"))
report.calculate()
print report.upload()
```
